```python
import string
S_P=string.punctuation
print(S_P)
```

```python
def remove_punctuation(data_ne):
    text_non_punctuation="".join([char for char in data_ne if char not in S_P])
    return text_non_punctuation

data_new['clean_text_1']=data_new['v2'].apply(lambda x: remove_punctuation(x))
data_new[0:10]
```

```python
import re 
def tokenise(text):
    tokens= re.split('\W+', text)
    return tokens
data_new['clean_tokenised']=data_new['clean_text_1'].apply(lambda x : tokenise(x.lower()))
data_new[0:10]
```

```python
from nltk.corpus import stopwords
stopWords=set(stopwords.words('english'))
#stopWords = nltk.corpus.stopwords.words('english')
def remove_stopwords(datta):
    text=[word for word in datta if word not in stopWords]
    return text
data_new['clean_nonpunc_nostop']=data_new['clean_tokenised'].apply(lambda x: remove_stopwords(x))
data_new[0:10]
```

```python
#wn=nltk.WordNetLemmatizer()
from nltk.stem import WordNetLemmatizer

wn = WordNetLemmatizer()
def lemmatizing(clean_nonpunc_nostop_text):
    text=[wn.lemmatize(word) for word in clean_nonpunc_nostop_text]
    return text
data_new['lemmatized_text']= data_new['clean_nonpunc_nostop'].apply(lambda x : lemmatizing(x))
data_new[0:20]
```
